### Делаем обёртки - свой API

```python
from collections import namedtuple
```

```python
Point = namedtuple('Point', ['x', 'y'])
Size  = namedtuple('Size', ['w', 'h'])

class Box(namedtuple('Box', ['x', 'y', 'w', 'h'], defaults=(0,0,1,1))):
    def __getattr__(self, key):
        if key in ('point', 'p', 'xy'):
            return Point(self.x, self.y)
        if key in ('size', 'wh'):
            return Size(self.w, self.h)
        return super().__getattr__(key)
        
    def __contains__(self, other):
        if isinstance(other, Box) or len(other) == 4 and (other := Box(other)):
            return (
                self.x <= other.x and
                self.y <= other.y and
                self.x + self.w > other.x + other.w and
                self.y + self.h > other.y + other.h
            )
        if isinstance(other, Point) or len(other) == 2 and (other := Point(other)):
            return (
                self.x <= other.x and
                self.y <= other.y and
                self.x + self.w > other.x and
                self.y + self.h > other.y
            )
        
```

```python
class ExpandableCell:
    def __init__(self, xlrd_cell, box:Box=None, sheet:Sheet=None, referred_from:Point = None):
        self.xlrd_cell = xlrd_cell
        self.box = box
        self.sheet = sheet
        self.referred_from = referred_from
        
    def __getattr__(self, key):
        if key in ('value', 'ctype', 'xf_index'):
            return getattr(self.xlrd_cell, key)
        if key in ('format'):
            idx = cell.xf_index
            if idx is not None:
                xf = self.sheet.xlrd_sheet.book.xf_list[idx]
                return xf
            else:
                raise ValueError('cell does not contain format data!')
        if key in ('border'):
            return self.format.border
        return super().__getattr__(key)
        ### return self.xlrd_cell.__getattr__(key)
        
    def __repr__(self):
        return '[{}]: {}'.format(self.box, repr(self.xlrd_cell))
```
